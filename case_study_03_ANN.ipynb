### DATA PREPROCESSING

```python
# Feature Extraction
x = data.drop('price_range', axis=1)
y = data['price_range']
```

```python
# splitting data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Data standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)   # Fit and transform the training set
x_test = scaler.transform(x_test)         # Transform the testing set using the parameters from the training set
```

### BUILDING NEURAL NETWOK MODEL

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
def build_model(rate):
    model = Sequential()
    model.add(Dense(20, input_dim = 20, activation='relu'))
    model.add(Dropout(rate))

    model.add(Dense(20, activation = "relu"))
    model.add(Dropout(rate))

    model.add(Dense(4, activation='softmax'))  # 4 classes for price_range

    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam',
                  metrics=['accuracy'])
    '''Here we are dealing with multiple classes (four classes for price_range),
    and the labels are integers (0, 1, 2, 3). Therefore, sparse_categorical_crossentropy
    is the appropriate choice.{It is used when  labels are integers} '''

    return model
```

```python
seed = 1
np.random.seed(seed)
```

```python
!pip install scikeras
from scikeras.wrappers import KerasClassifier
```

```python
# Wrap the Keras model using KerasClassifier
model = KerasClassifier(build_fn = build_model, verbose = 0)
```

```python
rate = [0, 0.2, 0.4]

epochs = [35, 50]

batch_size = [20]
```

```python
# Define the parameter grid
param_grid = dict(model__rate = rate , epochs = epochs, batch_size = batch_size)
```

```python
from sklearn.model_selection import GridSearchCV
# Use GridSearchCV to find the best combination of hyperparameters
grid_search = GridSearchCV(estimator = model, param_grid = param_grid, cv = 5)
```

```python
# Fit the model with the training data
grid_result = grid_search.fit(x_train, y_train)
```

```python
# Print the best parameters and corresponding accuracy
print(f"Best Parameters: {grid_result.best_params_}")
print(f"Best Accuracy: {grid_result.best_score_}")
```

```python
# Get the best model
best_model = grid_result.best_estimator_
```

```python
# Evaluate the best model on the test set
test_accuracy = best_model.score(x_test, y_test)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")
```
